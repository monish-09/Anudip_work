```python
# find out greatest digit present in given number
# enter user input
n=int(input("enter a number: "))
# given input convert str
s_n = str(n)

greatest_digit = -1
for i in s_n:
    digit = int(i)
    if digit > greatest_digit:
        greatest_digit = digit
print("the greatest digit is: ",greatest_digit)
```

```python
n=int(input("enter number: "))
g_n = 0
while n>0:
  digit = n%10
  if n>g_n:
    g_n = digit
  n = int(n//10)
print("the greatest digit is: ",g_n)
```
